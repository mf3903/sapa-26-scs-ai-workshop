```python
# Step 1: Check your environment
import sys
from importlib.metadata import version

print(f"Python: {sys.version}")
print(f"anthropic: {version('anthropic')}")
print(f"langgraph: {version('langgraph')}")
print(f"langchain: {version('langchain')}")
print("\nAll dependencies loaded successfully!")
```

```python
# Step 2: Load API key
import os
from dotenv import load_dotenv

load_dotenv()  # reads .env if present

api_key = os.getenv("ANTHROPIC_API_KEY")
if not api_key:
    raise ValueError("ANTHROPIC_API_KEY not set. See prerequisites above.")
print("API key loaded.")
```

```python
# Step 3: Minimal single-node LangGraph agent
from typing import TypedDict
from langgraph.graph import StateGraph, END
from langchain_anthropic import ChatAnthropic
from langchain_core.messages import HumanMessage

# --- State definition ---
class AgentState(TypedDict):
    messages: list
    response: str

# --- LLM ---
llm = ChatAnthropic(model="claude-haiku-4-5-20251001", anthropic_api_key=api_key)

# --- Node ---
def call_llm(state: AgentState) -> AgentState:
    response = llm.invoke(state["messages"])
    return {"messages": state["messages"], "response": response.content}

# --- Graph ---
graph = StateGraph(AgentState)
graph.add_node("llm", call_llm)
graph.set_entry_point("llm")
graph.add_edge("llm", END)
app = graph.compile()

print("Agent compiled successfully!")
```

```python
# Step 4: Run it!
result = app.invoke({
    "messages": [HumanMessage(content="In one sentence, what is a multi-agent system?")],
    "response": ""
})

print("Agent response:")
print(result["response"])
```
